```python
import pandas as pd
white_house = pd.read_csv("2015_white_house.csv")
status = white_house["Status"].unique().tolist()
status
```

```python
# figure out the relationship between status and salary
import numpy as np
status_salary = white_house.pivot_table(index="Status", values="Salary", aggfunc=np.mean)
total_stat = {}
for stat in white_house["Status"]:
    if stat not in total_stat:
        total_stat[stat] = 1
    else:
        total_stat[stat] += 1
```

```python
title = white_house["Position Title"]
```

```python
title_length = [len(tit.split(" ")) for tit in title]
```

```python
white_house["Title_length"] = title_length
```

```python
# figure out the relationships between salary and the length of title
import numpy as np
salary_titlen = white_house.pivot_table(index="Title_length", values="Salary", aggfunc=np.mean)
salary_titlen
```

```python
# the total salary paid for all staff in white house per annum
total_salary = white_house["Salary"].sum()
total_salary
```

```python
# find the highest and lowest salary and corresponding people
highest_salary = white_house["Salary"].max()
lowest_salary = white_house["Salary"].min()
print(highest_salary)
print(lowest_salary)
print(white_house[white_house["Salary"] == highest_salary])
print(white_house[white_house["Salary"] == lowest_salary])
```

```python
# count the frequencies of words showing up in titles 
title = white_house["Position Title"]
words_count = {}
for tit in title:
    for word in tit.split(" "):
        if word not in words_count:
            words_count[word] = 1
        else:
            words_count[word] += 1
            
max(words_count, key=words_count.get)
```

```python
words_count["THE"]
```

```python
import operator
sorted_wordscount = sorted(words_count.items(), key=operator.itemgetter(1))
```

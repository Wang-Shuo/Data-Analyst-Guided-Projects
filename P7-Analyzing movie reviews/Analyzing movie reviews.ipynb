```python
import pandas as pd

movies = pd.read_csv("fandango_score_comparison.csv")

movies
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.hist(movies["Fandango_Stars"])
plt.show()
plt.hist(movies["Metacritic_norm_round"])
plt.show()
```

```python
import numpy as np

F_mean = movies["Fandango_Stars"].mean()
M_mean = movies["Metacritic_norm_round"].mean()

F_median = movies["Fandango_Stars"].median()
M_median = movies["Metacritic_norm_round"].median()

F_std = np.std(movies["Fandango_Stars"])
M_std = np.std(movies["Metacritic_norm_round"])

print(F_mean,M_mean)
print(F_median,M_median)
print(F_std,M_std)
```

```python
plt.scatter(movies["Fandango_Stars"], movies["Metacritic_norm_round"])
plt.show()
```

```python
movies["fm_diff"] = np.abs(movies["Metacritic_norm_round"] - movies["Fandango_Stars"])
```

```python
movies.sort("fm_diff",ascending=False).head(5)
```

```python
from scipy.stats import pearsonr

r_value, p_value = pearsonr(movies["Metacritic_norm_round"], movies["Fandango_Stars"])
r_value
```

```python
from scipy.stats import linregress

slope,intercept,r_value,p_value,stderr = linregress(movies["Metacritic_norm_round"], movies["Fandango_Stars"])
```

```python
F_pred_1 = slope * 3 + intercept
F_pred_2 = slope * 4 + intercept
F_pred_1, F_pred_2
```

```python
x = [3.0, 4.0]
y = [F_pred_1, F_pred_2]
plt.plot(x, y)
plt.show()
```
